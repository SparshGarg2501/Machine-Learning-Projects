```python
import pandas as pd
import numpy as np
data=np.genfromtxt("0000000000002419_training_ccpp_x_y_train.csv",delimiter=",")
data.shape
x_train=data[:,0:4]
y_train=data[:,4:5]
from sklearn import preprocessing
scaler=preprocessing.StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
data[:,0:4]=x_train
data[:,4:5]=y_train
M=len(data)
print(M)
```

```python
def step_gradient(data,learning_rate,m):
    n=data.shape[1]
    m_slope=np.zeros(n)
    M=len(data)
    for i in range(M):
        x=data[i,0:4]
        x=np.append(x,1)
        y=data[i,4:5]
        for j in range(n):
            m_slope[j]=m_slope[j]+((-2/M)*(y-sum(m*x))*x[j])
        m=m-(learning_rate*m_slope)
    return m
```

```python
def cost(data,m):
    total_cost=0
    M=len(data)
    n=data.shape[1]
    for i in range(M):
        x=data[i,0:4]
        x=np.append(x,1)
        y=data[i,4:5]
        total_cost=total_cost+(1/M)*((y-(m*x).sum())**2)
    return total_cost
```

```python
def gd(data,learning_rate,iterations):
    n=data.shape[1]
    m=np.zeros(n)
    for i in range(iterations):
        m=step_gradient(data,learning_rate,m)
        print(i,"cost:",cost(data,m))
    return m
```

```python
def run():
    learning_rate=0.0001
    iterations=100
    m =gd(data,learning_rate,iterations)
    return m[0:4],m[4]
```

```python
m,c = run()
```

```python
def predict(x, m, c):
    N = x.shape[0]
    Y = np.zeros(N)
    for i in range(N):
        X = x[i,:]
        Y[i]=(m * X).sum() + c
    return Y
```

```python
x_test=np.genfromtxt("0000000000002419_test_ccpp_x_test.csv",delimiter=",")
x_test=scaler.transform(x_test)
predictions=predict(x_test, m, c)

predictions=pd.DataFrame(predictions)
predictions.columns=['predictions']
predictions
```
