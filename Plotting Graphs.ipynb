```python
import numpy as np
#color, marker, linewidth, labels for axis, datasets, title, axis, grid, text
#x = np.array([1,2,3,4])
x = np.arange(0,5,0.1)
y = x **3
y2 = x**2
#plt.plot(x,y,"black", marker="s")
plt.plot(x,y,"black", linewidth=5, label = "x^3")

plt.plot(x,y2,color="red", linewidth=5, label = "x^2")
plt.ylabel("X^3")
plt.xlabel("x")
plt.legend()
plt.grid()
plt.text(2,80,"text", fontsize=12)
plt.title("Matlpotlib demo")
plt.show()
```

```python
#pie graphs
labels = ["A","B", "C", "D"]
sizes = [3,4,6,2]
colors = ["blue", "green", "red", "yellow"]
explode=[0.1,0.2,0,0]
plt.title("Split among classes")
plt.pie(sizes, colors=colors,explode=explode, labels=labels,autopct="%.2f%%", startangle=100)
plt.show()
```
